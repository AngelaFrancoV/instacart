```python
# Step 5 - Expenditure depending on the hour of the day
df_3 = small[['order_hour_of_day','prices']]
line_hour = sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices')
```

```python
# Step 6 - exploratory analysis of customer demographics
df_4 = small[['Age','n_dependants']]
line_demographic = sns.lineplot(data = df_4, x = 'Age',y = 'n_dependants')
```

```python
# The line chart shows that there's no connection between age and number of dependants. 
People in their 20s, as well as people in their 60s and 70s have the higher number of dependants
```

```python
# Step 7 - Connection between age and income
sns.scatterplot(x = 'Age', y = 'income',data = project, alpha=0.5)
```

```python
# The scatter plot shows that some customers from 40 - 80 years old have a higher income than those from 20 -40 years old
```
