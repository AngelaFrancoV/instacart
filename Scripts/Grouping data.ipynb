```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
#Import dataframe
ords_prods_merge = pd.read_pickle(r'C:/Users/ab-fv/OneDrive/Documentos/Instacart Basket Analysis/02 Data/Prepared Data/ords_prods_new_columns.pkl')
```

```python
# Step 2 Aggregated mean of order number
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Step 3 - Result analysis
The departments mean of the entire dataframe is slightly higher than the one from the subset but it's not very different
```

```python
# Step 4 - Creating loyalty flag
```

```python
# Create max_order column 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Check
ords_prods_merge.head(5)
```

```python
# Step 5
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
The new customers tend to spend more money in orders and the loyal customers the least
```

```python
# Step 6 - Spending flag
  # Create average column
ords_prods_merge['avg_price'] = ords_prods_merge.groupby('user_id')['prices'].mean()
ords_prods_merge.head(5)
```

```python
# Group
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.head(5)
```

```python
# Step 7 - Frequecy flag
```

```python
# Create avg days since prior order
ords_prods_merge['avg_days'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].mean()
```

```python
# Group
ords_prods_merge.loc[ords_prods_merge['avg_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['avg_days'] > 10) & (ords_prods_merge['avg_days'] <= 20), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['avg_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
ords_prods_merge[['user_id', 'avg_days', 'frequency_flag']].head(20)
```

```python
# Step 9 - Export file in pickle format
ords_prods_merge.to_pickle(r'C:/Users/ab-fv/OneDrive/Documentos/Instacart Basket Analysis/02 Data/Prepared Data/ords_prods_flags.pkl')
```
